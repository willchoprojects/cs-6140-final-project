## Data Cleanup

```python
OUTPUT_COLS = ['COVID_Mortality_R']
```

```python
POP_DATA_INPUT_COL_NAMES = [
    'POP_DENS',
    'POP 0-14',
    'POP 15-29',
    'POP 30-44',
    'POP 45-59',
    'POP_GROWTH',
    'HOUSEHOLDS',
    'DWELLINGS',
]
```

```python
POP_DATA_INCLUDED_COL_NAMES = OUTPUT_COLS + POP_DATA_INPUT_COL_NAMES
```

```python
POP_ENV_DATA_INCLUDED_COL_NAMES = POP_DATA_INCLUDED_COL_NAMES + [
    'SURFACE AREA',
    'NTL_avg',
    'NO2_avg',
    'WIND_Avg',
    'Temp_mean_over12yrs',
    'PRESSURE_avg',
    'PRECIPITATION',
    'SOLAR_RAD_avg',
    'LAI_avg',
]
```

```python
cleaned_data_df = raw_data_df.drop('-').dropna(subset=POP_ENV_DATA_INCLUDED_COL_NAMES)
```

```python
cleaned_data_df.head()
```

```python
print(f'The number of data points is: {len(cleaned_data_df)}')
```

```python
pop_data_df = cleaned_data_df[POP_DATA_INCLUDED_COL_NAMES]
```

```python
pop_data_df.head()
```

```python
print('Number of NaNs by column')

for col_name in pop_data_df.columns:
    print(f'{col_name}: {pop_data_df[col_name].isna().sum()}')
```

```python
print(f'The number of data points is: {len(pop_data_df)}')
```

```python
pop_env_data_df = cleaned_data_df[POP_ENV_DATA_INCLUDED_COL_NAMES]
```

```python
pop_env_data_df.head()
```

```python
print('Number of NaNs by column')

for col_name in pop_env_data_df.columns:
    print(f'{col_name}: {pop_env_data_df[col_name].isna().sum()}')
```

```python
print(f'The number of data points is: {len(pop_env_data_df)}')
```

## Save Data Frames

```python
POP_DATA_DF_FILE_NAME = 'pop_data_df'
POP_ENV_DATA_DF_FILE_NAME = 'pop_env_data_df'
EXTENSION = '.csv'
```

```python
pop_data_df.to_csv(f'{BASE_PATH}/{POP_DATA_DF_FILE_NAME}{EXTENSION}')
pop_env_data_df.to_csv(f'{BASE_PATH}/{POP_ENV_DATA_DF_FILE_NAME}{EXTENSION}')
```

## Implement Train/Test Splits

```python
TEST_SIZE = 0.33
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = pop_env_data_df.drop(OUTPUT_COLS, axis=1)
Y = pop_env_data_df[OUTPUT_COLS]
```

```python
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=TEST_SIZE)
```

```python
pop_data_df_train_X = train_X[POP_DATA_INPUT_COL_NAMES]
pop_data_df_test_X = test_X[POP_DATA_INPUT_COL_NAMES]
pop_data_df_train_Y = train_Y
pop_data_df_test_Y = test_Y
```

```python
pop_data_df_train_X.to_csv(f'{BASE_PATH}/{POP_DATA_DF_FILE_NAME}_train_X{EXTENSION}')
pop_data_df_test_X.to_csv(f'{BASE_PATH}/{POP_DATA_DF_FILE_NAME}_test_X{EXTENSION}')
pop_data_df_train_Y.to_csv(f'{BASE_PATH}/{POP_DATA_DF_FILE_NAME}_train_Y{EXTENSION}')
pop_data_df_test_Y.to_csv(f'{BASE_PATH}/{POP_DATA_DF_FILE_NAME}_test_Y{EXTENSION}')
```

```python
pop_env_data_df_train_X = train_X
pop_env_data_df_test_X = test_X
pop_env_data_df_train_Y = train_Y
pop_env_data_df_test_Y = test_Y
```

```python
pop_env_data_df_train_X.to_csv(f'{BASE_PATH}/{POP_ENV_DATA_DF_FILE_NAME}_train_X{EXTENSION}')
pop_env_data_df_test_X.to_csv(f'{BASE_PATH}/{POP_ENV_DATA_DF_FILE_NAME}_test_X{EXTENSION}')
pop_env_data_df_train_Y.to_csv(f'{BASE_PATH}/{POP_ENV_DATA_DF_FILE_NAME}_train_Y{EXTENSION}')
pop_env_data_df_test_Y.to_csv(f'{BASE_PATH}/{POP_ENV_DATA_DF_FILE_NAME}_test_Y{EXTENSION}')
```
